# Rearrange files to ImageFolder format
The PyTorch ```ImageFolder``` class expects images to be within directories, with one for each class of interest to work. For example:
- root/dog/xxx.png
- root/dog/xxy.png
- root/dog/xxz.png

- root/cat/123.png
- root/cat/nsdf3.png
- root/cat/asd932_.png

This notebook rearranges files to match the file hierarchy required by ```ImageFolder```.

````python
import os

def rearrange(folder_path):
    for filename in os.listdir(folder_path):
        old_path = folder_path + filename
        new_path = ''
        
        if os.path.isfile(old_path):
            if (filename[5] == '0'):
                new_path = folder_path + '0/' + filename

            elif (filename[5] == '1'):
                new_path = folder_path + '1/' + filename
            
            os.rename(old_path, new_path)
            print(f"Moved {old_path} to {new_path}")

rearrange(r"../dataset-resnet/train/")
rearrange(r"../dataset-resnet/test/")
````
