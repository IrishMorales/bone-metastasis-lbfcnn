# Max Normalization
This notebook demonstrates the use of max normalization. Normalization is done using the formula x/max_value, where:
- x - pixel value
- max_value - maximum known value in the image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pprint import pprint
```

```python
# Read image and convert to numpy array
input_img = cv2.imread('sample_input/arf.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(input_img, cmap='gray', vmin=0, vmax=255)
```

```python
# Inspect values in numpy array
print(input_img)
```

```python
# Apply max normalization to each pixel value
normalized_img = np.array([[pixel/255 for pixel in row] for row in input_img])
print(normalized_img)
```

```python
plt.imshow(normalized_img, cmap='gray', vmin=0, vmax=1)
```

```python
# Plot histogram of values before and after normalization to inspect changes
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
axs[0].hist(input_img.flatten())
axs[1].hist(normalized_img.flatten())
plt.show()
```
